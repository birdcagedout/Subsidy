```python
# '주소'가 '노원구'인 것들만 추출하되 '노원구'만으로 검색하면 안 됨 ==> 타구 주소인데 도로명에 '추노원구단1길' 같은 것이 없는지 확인
nowon_cars = cars[cars['주소'].str.contains('서울특별시 노원구|서울시 노원구')]
nowon_cars
```

```python
# 행 index를 0부터 세팅하고 + 이전 인덱스값은 모두 지움
nowon_cars = nowon_cars.reset_index(drop=True)
nowon_cars
```

```python
# 4개 열(총액, 국비, 도비, 시비, 추가) 천단위 콤마 있는지 확인 ==> 있다면 제거(find(',')!=-1 대신  replace(','. ''))해서 원래 df에 덧빵
comma_df = nowon_cars[['신청대수', '총액', '국비', '도비', '시비', '추가']].apply(lambda x: x.astype(str).str.find(',') != -1)
comma_df.value_counts()
```

```python
# 도비는 모두 0이므로 열을 지운다
nowon_cars = nowon_cars.drop(['도비'], axis=1)
nowon_cars
```

```python
# [총액 = 국비 + 시비 + 추가] 금액이 맞는지 확인 ==> 나오는 게 없어야 정상
nowon_cars[pd.to_numeric(nowon_cars['총액']) != pd.to_numeric(nowon_cars['국비']) + pd.to_numeric(nowon_cars['시비']) + pd.to_numeric(nowon_cars['추가'])]
```

```python
# '시비'열에 '추가'열을 더해 '시비'열에 저장
# 열은 하나의 type으로만 저장가능: int형과 float형이 섞여 있을 때 모두 float로 형변환 되는 것을 방지하기 위해 ==> 금액 확인 후 int형/float형 구분 후 str로 저장
for i in range(nowon_cars.shape[0]):
    sibi_extra_sum = float(nowon_cars.loc[i, '시비']) + float(nowon_cars.loc[i, '추가'])
    if sibi_extra_sum == int(sibi_extra_sum):
        nowon_cars.loc[i, '시비'] = str(int(sibi_extra_sum))
    else:
        nowon_cars.loc[i, '시비'] = str(sibi_extra_sum)

# '추가'열 삭제
nowon_cars = nowon_cars.drop(['추가'], axis=1)

# [총액 = 국비 + 시비 + 추가] 금액이 맞는지 확인 ==> 나오는 게 없어야 정상
nowon_cars[pd.to_numeric(nowon_cars['총액']) != pd.to_numeric(nowon_cars['국비']) + pd.to_numeric(nowon_cars['시비'])]
```

### 한줄에 여러대(멀티케이스) 존재하는 경우 ==> 처리 시작

```python
# 한줄 여러대(멀티케이스)가 몇 건 있는지 저장
multicase_df = nowon_cars[nowon_cars['신청대수'] != '1'].copy()
multicase_count = multicase_df.shape[0]
multicase_count
```

```python
# 멀티케이스의 각 row = Series
multicase_df.iloc[0]
```

```python
for k in range(multicase_count):
    # loc(index이름) 찾기
    print(f"loc = {multicase_df.index[k]}")

    # iloc(index숫자) 찾기 : 위에서 720(loc)이 나왔다면 이것이 몇번째 인덱스(29)인지 확인한다.
    print(f"iloc = {nowon_cars.index.get_indexer(multicase_df.index)[k]}")
```

```python
# 만약 나온 게 있다면 ==> 1개 행에 신청 대수가 2대 이상인 경우 처리 II
duplicated_df_list = []
for j in range(0, multicase_count):
    
    original_row = multicase_df.iloc[j].copy()    # 기준 행 = Series
    original_row_count = original_row['신청대수']
    original_row_iloc = nowon_cars.index.get_indexer(multicase_df.index)[j]

    nums = []
    vins = []
    regs = []

    nums = original_row['차량번호'].split()
    vins = original_row['차대번호'].split()
    regs = original_row['차량등록일'].split()

    original_row['신청대수'] = int(original_row['신청대수'] / original_row_count)
    original_row['총액'] = int(original_row['총액'] / original_row_count)
    original_row['국비'] = int(original_row['국비'] / original_row_count)
    original_row['시비'] = int(original_row['시비'] / original_row_count)

    original_row['차량번호'] = nums[0]
    original_row['차대번호'] = vins[0]
    original_row['차량등록일'] = regs[0]


    case_processed_df = original_row
    for i in range(1, original_row_count):
        temp_row = original_row.copy()
        temp_row['차량번호'] = nums[i]
        temp_row['차대번호'] = vins[i]
        temp_row['차량등록일'] = regs[i]
        case_processed_df = pd.concat([case_processed_df, temp_row], axis=1, ignore_index=True)

    # 원본 행 + 분리된 행 = 멀티케이스 1건당 새로운 1개의 DataFrame
    case_processed_df = case_processed_df.T
    duplicated_df_list.append(case_processed_df)
    print(case_processed_df)
    
    # 원본에서 기준행 1개 지우기
    nowon_cars.drop(multicase_df.index[j], inplace=True)


# nowon_cars에 삽입    
for m in range(0, multicase_count):
    nowon_cars = pd.concat([nowon_cars, duplicated_df_list[m]], axis=0, ignore_index=True)

nowon_cars
```

```python
# 신청 대수가 2대 이상인 경우 나오는지 검증 ==> 나오는 것이 없어야 정상 (없으면 멀티케이스 처리 끝)
nowon_cars[nowon_cars['신청대수'] != 1]
```
