```python
import pandas as pd
```

```python
# 첫번째 시트 : 전기차
car1 = pd.read_excel("1.xlsx", dtype="string", header=4, sheet_name=0)
car1
```

```python
# 두번째 시트 : 수소차
car2 = pd.read_excel("1.xlsx", dtype="string", header=4, sheet_name=1)
car2
```

```python
cars = pd.concat([car1, car2], ignore_index=True)
cars
```

```python
# 열 이름 가져오기
cars.keys()
```

```python
# 열 이름을 알기쉽게 변경
cars.columns = ['차종구분', '신청번호', '지급일자', '차종', '신청대수', '신청자명', '주소',
       '총액', '국비', '도비', '시비', '추가', '차량번호',
       '차대번호', '차량출고일', '차량등록일']

# 불필요한 열은 삭제
cars = cars.drop(['차종구분', '신청번호', '차량출고일'], axis=1)
cars
```

```python
# '주소'가 '노원구'인 것들만 추출 ==> 타구 주소인데 도로명에 '노원구세군1길' 같은 것이 없는지 확인
# nowon_cars = cars[cars['주소'].str.contains('노원구')]

nowon_cars = cars[cars['주소'].str.contains('서울특별시 노원구|서울시 노원구')]
nowon_cars
```

```python
# 신청대수가 2대 이상인 경우 걸러내기(엑셀 행 1개에 여러대인 경우 1/N)
# ==> 아무것도 나오지 않아야 정상
nowon_cars[nowon_cars['신청대수'] != '1']
```

```python
# 도비가 0이 아닌 비정상적인 경우를 걸러내기(서울시는 도비가 없음)
# ==> 아무것도 나오지 않아야 정상
nowon_cars[nowon_cars['도비'] != '0']
```

```python
# 도비는 모두 0이므로 열을 지운다
nowon_cars = nowon_cars.drop(['도비'], axis=1)
nowon_cars
```

```python
# 이제 확인이 끝났으므로 불필요한 '차종', '신청대수', '주소' 열을 지운다.
nowon_cars = nowon_cars.drop(['차종', '신청대수', '주소'], axis=1)
nowon_cars
```

```python
# 4개 열(총액, 국비, 시비, 추가) 문자열 ==> 정수형으로 변환 + 천단위 콤마 제거
nowon_cars[['총액', '국비', '시비', '추가']] = nowon_cars[['총액', '국비', '시비', '추가']].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',',''), errors='coerce'))
nowon_cars.dtypes
```

```python
nowon_cars
```

```python
# 총액 = 국비 + 시비 + 추가 ==> 금액이 맞는지 확인 ==> 나오는 게 없어야 정상
nowon_cars[nowon_cars['총액'] != nowon_cars['국비'] + nowon_cars['시비'] + nowon_cars['추가']]
```

```python
# 특기사항이 화면에 모두 출력될 수 있도록 열의 최대출력 길이 변경
pd.set_option('display.max_colwidth', 200)

# 인덱스(행 이름) 0부터 정렬 + dataframe 이름 변경
df = nowon_cars.reset_index(drop=True)
df
```

```python
# 서울시 공문번호 입력
doc_num = "서울시 기후변화대응과-24031(2022.06.17.)"

# 특기사항 입력
# 서울특별시, 전기차 구매보조금 339만원(국비 264만원, 시비 75만원, 지급일자 2022.5.24.) 서울시 기후변화대응과-24031(2022.06.17.)
df['특기사항'] = "서울특별시, 전기차 구매보조금 " + (df['총액']).astype('str') + "만원 (국비 " + (df['국비']).astype('str') + "만원, 시비 " + (df['시비'] + df['추가']).astype('str') + "만원, 지급일자 " + df['지급일자'] + ") " + doc_num
df
```

```python
# pip install XlsxWriter

writer = pd.ExcelWriter("2022_07_보조금.xlsx") 
df.to_excel(writer, sheet_name='보조금', index=False)

for column in df:
    column_length = max(df[column].astype(str).map(len).max(), len(column))
    col_idx = df.columns.get_loc(column)
    writer.sheets['보조금'].set_column(col_idx, col_idx, column_length*1.3)    # 한글 때문에 X1.3으로 열너비 늘렸다.

writer.save()
```

### 전처리로 만들어진 xlsx 파일 ==> 후처리에서 특기사항 입력
