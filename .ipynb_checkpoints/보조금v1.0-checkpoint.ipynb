```python
import time
import pyperclip
import pandas as pd
from pyautogui import *
```

```python
car_list = pd.read_excel("보조금입력(12월).xlsx", dtype="string")
```

```python
car_list = car_list[['차량번호', '특기사항']]
```

```python
len(car_list)
```

```python
car_list
```

### 1. 보조금 ==> 특기사항 입력

```python
for i in range(32, len(car_list)):
    
    regnum = car_list['차량번호'][i]
    memo = car_list['특기사항'][i]
    
    # 초기화
    pos = locateCenterOnScreen("init.png", grayscale=True)
    while pos is None:
        print(f"[오류] 초기화 버튼 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("init.png", grayscale=True)
    moveTo(pos)
    click()
    time.sleep(1)
    
    # 등록번호 입력
    pos = locateCenterOnScreen("regnum.png", grayscale=True)
    while pos is None:
        print(f"[오류] 등록번호 입력칸 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("regnum.png", grayscale=True)
#     moveTo(pos)
#     moveRel(100, 0)
    pos = Point(x=pos.x+100, y=pos.y)
    moveTo(pos)
    click()
    pyperclip.copy(regnum)
    hotkey('ctrl', 'v')
    
    # 조회 클릭
    pos = locateCenterOnScreen("search.png", grayscale=True)
    while pos is None:
        print(f"[오류] 조회 버튼 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("search.png", grayscale=True)
    moveTo(pos)
    click()
    time.sleep(1)
    
    # 특기사항 입력
    pos = locateCenterOnScreen("memo.png", grayscale=True)
    while pos is None:
        print(f"[오류] 특기사항 입력칸 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("memo.png", grayscale=True)
#     moveTo(pos)
#     moveRel(0, 50)
    pos = Point(x=pos.x, y=pos.y+50)
    moveTo(pos)
    click()
    pyperclip.copy(memo)
    hotkey('ctrl', 'v')
    
    # 저장 클릭
    pos = locateCenterOnScreen("save.png", grayscale=True)
    while pos is None:
        print(f"[오류] 저장 버튼 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("save.png")
    moveTo(pos)
    click()
    time.sleep(1)
    
    # 저장 완료되었습니다. "확인" 클릭
    pos = locateCenterOnScreen("complete.png")
    while pos is None:
        print(f"[오류] 저장 완료 후 확인 버튼 찾기 ==> 현재 index={i}")
        time.sleep(1)
        pos = locateCenterOnScreen("complete.png")
#     moveTo(pos)
#     moveRel(170, 120)
    pos = Point(x=pos.x+170, y=pos.y+120)
    moveTo(pos)
    click()
    time.sleep(1)
```

### 2. 특기사항 추출 ==> 엑셀 파일 내용과 비교 및 검증

```python
for i in range(len(car_list)):
    
    regnum = car_list['차량번호'][i]
    memo = car_list['특기사항'][i]
    
    # 초기화
    pos = locateCenterOnScreen("init.png")
    while pos is None:
        print(f"[오류] 초기화 버튼 찾기 ==> 현재 index={i}")
        time.sleep(0.5)
        pos = locateCenterOnScreen("init.png")
    moveTo(pos)
    click()
    time.sleep(0.5)
    
    # 등록번호 입력
    pos = locateCenterOnScreen("regnum.png")
    while pos is None:
        print(f"[오류] 등록번호 입력칸 찾기 ==> 현재 index={i}")
        time.sleep(0.5)
        pos = locateCenterOnScreen("regnum.png")
#     moveTo(pos)
#     moveRel(100, 0)
    pos = Point(x=pos.x+100, y=pos.y)
    moveTo(pos)
    click()
    pyperclip.copy(regnum)
    hotkey('ctrl', 'v')
    
    # 조회 클릭
    pos = locateCenterOnScreen("search.png")
    while pos is None:
        print(f"[오류] 조회 버튼 찾기 ==> 현재 index={i}")
        time.sleep(0.5)
        pos = locateCenterOnScreen("search.png")
    moveTo(pos)
    click()
    time.sleep(0.5)
    
    # 오늘 입력한 특기사항 추출
    pos = locateCenterOnScreen("list_item.png", grayscale=True)
    while pos is None:
        print(f"[오류] 오늘 금방 입력한 특기사항 리스트 아이템  ==> 현재 index={i}")
        time.sleep(0.5)
        pos = locateCenterOnScreen("list_item.png")
    moveTo(pos)
    doubleClick()

    pos = locateCenterOnScreen("memo.png")
    while pos is None:
        print(f"[오류] 특기사항 입력칸 찾기 ==> 현재 index={i}")
        time.sleep(0.5)
        pos = locateCenterOnScreen("memo.png")
#     moveTo(pos)
#     moveRel(0, 50)
    pos = Point(x=pos.x, y=pos.y+50)
    moveTo(pos)
    click()
    hotkey('ctrl', 'a')
    hotkey('ctrl', 'c')

    contents = pyperclip.paste()
    if contents == memo:
        print(f"[확인 {(i+1):>2}/{len(car_list):>3}] 등록번호: {regnum} \t =====> 특기사항 입력 확인 OK")
    else:
        print(f"[확인 {(i+1):>2}/{len(car_list):>3}] 등록번호: {regnum} \t =====> 특기사항 입력 오류 ㅠㅠ")    
    time.sleep(1)
	
print("특기사항 입력 확인 끝!!")
```
